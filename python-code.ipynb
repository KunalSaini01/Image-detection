```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('images/ks.jfif')


plt.imshow(img)

cv.waitKey(0)    
```

```python

RGB_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(RGB_img)
```

```python
image = img
B, G, R = cv.split(image) 
# Corresponding channels are separated  
plt.title('blue'),plt.imshow(B) 
cv.waitKey(0) 
print("blue")
```

```python
plt.title('Green'),plt.imshow( G) 
cv.waitKey(0)
```

```python
plt.title('RED'),plt.imshow(R) 
cv.waitKey(0)
```

```python
edges = cv.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```

```python
blur = cv.GaussianBlur(img,(7,7),cv.BORDER_DEFAULT)
plt.title('GaussianBlur'),plt.imshow(blur)
```

```python
imgs = cv.imread('images/d.jfif')
```

```python
gray = cv.cvtColor(imgs, cv.COLOR_BGR2GRAY)
```

```python
face_cascade = cv.CascadeClassifier('haarcascade_frontalface_alt2.xml') 

# Detect faces 
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
```

```python
for (x, y, w, h) in faces: 
    cv.rectangle(imgs, (x, y), (x+w, y+h), (0, 0, 255), 2) 
    faces = imgs[y:y + h, x:x + w] 
    cv.imshow("face",faces) 
    cv.imwrite('face.jpg', faces)
```

```python

RGB_img = cv.cvtColor(imgs, cv.COLOR_BGR2RGB)
plt.title('detcted.jpg'),plt.imshow(RGB_img)

```

```python
cv.imwrite('detcted.jpg', imgs) 
cv.imshow('img', imgs) 
cv.waitKey()
```
